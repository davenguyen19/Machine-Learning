# Logistic Regression

You will implement logistic regression classifier.

Iris dataset with only 2 classes is provided to test the code.


```python
import numpy as np
import pandas as pd
import numpy.matlib
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# read dataset
df = pd.read_csv("pulsar_stars.csv")
print(df.head())

df = df.values
X = df[:,0:8]   # 100x4
y = df[:,8]     # 100x1
```

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print ('X_train dimension = {0}; X_test dimension = {1}'.format(X_train.shape, X_test.shape))

# Train model using training data
model = LogisticRegression().fit(X_train, y_train)
```

```python
# Test model with test data
y_hat = model.predict(X_test)
score = model.score(X_test, y_test)

for i in range (0, len(y_test)):
  print('y={0}, y_hat={1}'.format(y_test[i], y_hat[i]))

print('Average accuracy = {0:2f}'.format(score))
```
